```python
import tensorflow as tf

img_tf = tf.io.read_file('images/lenna.png')
img_tf = tf.image.decode_image(img_tf, channels = 3)
plt.imshow(img_tf)
```

# Create zero matrix and plot it

```python
import numpy as np
zeros = np.zeros([100,100])
zeros
```

```python
plt.imshow(zeros, cmap = 'gray')
```

```python
zeros[56,89] = 0.9
plt.imshow(zeros, cmap = 'gray')
```

```python
zeros[60] = 5
plt.imshow(zeros, cmap = 'gray')
```

```python
# uint8 => unsigned (no negative pixel is taken)
zeros=np.zeros([10000],dtype=np.uint8)
```

```python
for i in range(10000):
    zeros[i]=np.random.randint(0,255)
```

```python
zeros=zeros.reshape(100,100)
plt.imshow(zeros, cmap = 'gray')
```

# convert RGB -> BGR channels

```python
plt.subplot(121)
plt.imshow(img_plt)
plt.subplot(122)
plt.imshow(img_cv)
# when it is red in img=> green and vice versa
```

```python
# convert bgr->rgb ==> use COLOR_BGR2RGB

img_cv_new = cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB)

plt.subplot(121)
plt.imshow(img_plt)
plt.subplot(122)
plt.imshow(img_cv_new)
```

# memory taking by the image:

```python
import sys
sys.getsizeof(img_plt)      # gives number of bits
```

```python
sys.getsizeof(img_cv)     # size is small as compare to img_plt => dtype is uint8 for img_cv_new
```

```python
# check the dtype:
print(img_plt.dtype)

# convert float32 into float16 for img_plt

img_plt_new = img_plt.astype(np.float16)
img_plt_new.dtype
```

```python
sys.getsizeof(img_plt_new)   # still not small as img_cv_new
```
