# Problem statement:

for given input alphabet predict its next alphabet

eg, if input (A) => output(B)

```python
alphabets = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
# Create dict: index as key and alphabet as value:

int_to_char  = list(enumerate(alphabets))
int_to_char 
```

# Dictionary Compression method:

```python
# if we want alphabet as key and index as value

char_to_int = dict((v,i) for i, v in enumerate(alphabets))
char_to_int
```

# Data preparation

```python
seq_length=1   # len(input)
x=[]
y=[]

for i in range(len(alphabets)-seq_length):
    seq_in=[alphabets[i]]
    seq_out=alphabets[i+1]
    x +=[[char_to_int[seq_in[0]]]]
    y +=[char_to_int[seq_out]]
    print(seq_in,'->',seq_out)

    
```

```python
print(x,y)
```

# prepare the output variable

```python
from keras.utils import to_categorical
```

```python
#apply one hot encoding
y_new=to_categorical(y)
y_new
```

# prepare input variable


```python
import numpy as np
x =np.reshape(x,(25,1,1))
x
```

```python
x.shape
```

```python
# Normalizing the data using min-max method (as min = 0 ==> x / x<max>)
# X_new=input, y_new=output

x_new = x/25           # z is also consider here
x_new.shape
```

```python
y_new.shape
```

# create the model

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
model=Sequential()
model.add(Dense(25,input_shape=(1,1),activation='relu'))
model.add(LSTM(32))
model.add(Dense(26,activation='softmax'))
```

# compile the model

```python
model.compile(loss="categorical_crossentropy",optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x_new,y_new,epochs=1000,batch_size=1)
```

```python
model.evaluate(x_new,y_new,batch_size = 1)
```

# predict the character in sequence

```python
new = 'T'

# Get the int value of the alpha
new=char_to_int[new]

# reshape it
new=np.reshape(new,(1,1,1))

# Normalized it:
new = new/25

# Apply predict
pred=model.predict(new)

pred=np.argmax(pred)
int_to_char[pred]
```

#  multiple char input and one char output:

```python
seq_length = 3
x = []
y = []

for i in range(len(alphabets) - seq_length):
    seq_in = alphabets[i: i + seq_length]
    seq_out = alphabets[i + seq_length]
    x.append([char_to_int[char] for char in seq_in])
    y.append(char_to_int[seq_out])
    print(seq_in, '=>', seq_out)
```

# Train the algorithm for multiple input:

```python
x = np.reshape(x, (len(x), seq_length, 1))

# scaled it:
x = x/25

# build the model

model=Sequential()
model.add(Dense(25,input_shape=(3,1),activation='relu'))
model.add(LSTM(128))
model.add(Dense(26, activation='softmax'))
```

```python
# Apply one hot encoding:

y_new = to_categorical(y)
y_new.shape
```

```python
model.compile(loss="categorical_crossentropy", optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x, y_new, epochs=500, batch_size=1)
```

```python
model.evaluate(x, y_new, batch_size = 1)
```

```python
# predict:

new = 'HIJ'

# Get the int value of the alpha
val = np.array([[char_to_int[x]] for x in new])

# reshape it:
val = val.reshape(1,3,1)
val = val/ 25

res = model.predict(val)
res = np.argmax(res)
print(int_to_char[res])
```
