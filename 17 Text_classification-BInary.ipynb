# Prepare the Model for Vectorization: 

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
sent = 'Hello friends! How are you? I like Python Programming.'
```

# clean the data:

# 1) Apply Tokenize:

```python
token1 = word_tokenize(sent)
token1
```

# 2) Remove punctuation:

```python
token2 = [token for token in token1 if token.isalnum()]
token2
```

# 3) Remove the stopwords:

```python
swords = stopwords.words('english')

token3 = [token for token in token2 if token.lower() not in swords]
token3
```

# 4) Stemming:

```python
ps = PorterStemmer()

token4 = [ps.stem(token) for token in token3]
token4
```

# Try steps 1 to 4 in function:

```python
# Create one function:

def clean_text(sent):
        # apply tokenization:
    tokens1 = word_tokenize(sent)
    
    # remove punct:
    tokens2 = [token for token in  tokens1 if token.isalnum()]
    
    # remove stopwords:
    tokens3 = [token for token in tokens2 if token.lower() not in swords]
    
    # stemming:
    tokens4 = [ps.stem(token) for token in tokens3] 
    
    return tokens4

# call the function and pass the sent:
clean_text(sent)
    
```

```python
# check with another sent:

sent1 = 'It was the expected result on that occasion.'
clean_text(sent1)
```

```python
# by giving sent number:

clean_text(x.loc[400])
```

# pass all the corpus to the function:


```python
x.apply(clean_text)
```

# TF*IDF vectorizer

```python
# import class
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#create object
tfidf = TfidfVectorizer(analyzer=clean_text)

#apply
x_new = tfidf.fit_transform(x)
x_new

# 6854 features
```

# Cross-validation Process => spliting

```python
#import the function
from sklearn.model_selection import train_test_split

#seprate the training and testing data
x_train,x_test,y_train,y_test = train_test_split(x_new,y,random_state=0)
```

```python
tfidf.get_feature_names_out()
```

# Build the ML clasification model

```python
#import the class
from sklearn.svm import SVC

#create object
svm = SVC(kernel='poly',random_state=0)
```

```python
%%time
svm.fit(x_train,y_train)
```

# Performance Evaluation:

```python
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay, classification_report, confusion_matrix

y_pred = svm.predict(x_test)
print("The accuracy is: ",accuracy_score(y_test, y_pred))

# Confusion mat:
print(confusion_matrix(y_test, y_pred))

# confusion mat display:
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
