```python
from urllib.request import urlopen
```

```python
url = urlopen("https://en.wikipedia.org/wiki/Rajgad_Fort")
```

# read the url data:


```python
data=url.read()

# it will give us the html docs format data:
print(data)
```

```python
from bs4 import BeautifulSoup
```

```python
# create obj:

soup = BeautifulSoup(data, 'html.parser')
soup
```

```python
type(soup)
```

# Extract the plain text:

```python
text = soup.get_text(strip = True)
print(text)
```

# Top 3 words:

```python
from nltk.tokenize import word_tokenize
from nltk import FreqDist
```

```python
tokens = word_tokenize(text)
freq = FreqDist(tokens)

freq.plot(20)
```

# Remove the punctuation and stopwords:

```python
from nltk.corpus import stopwords

swords = stopwords.words('english')

# remove stopwords:
token1 = [token for token in tokens if token.lower() not in swords]

# remove punctuation:
token2 = [token.lower() for token in token1 if token.isalnum()]
```

```python
# PLot the graph:

freq = FreqDist(token2)
freq.plott(10)
```

# Another Example ==> Tiger

```python
from nltk.stem import PorterStemmer,WordNetLemmatizer

ps = PorterStemmer()
wnl = WordNetLemmatizer()


url = urlopen("https://en.wikipedia.org/wiki/Tiger")
data=url.read()

soup =BeautifulSoup(data,"html.parser")
text = soup.get_text(strip = True)

tokens= word_tokenize(text)
tokens1 = [wnl.lemmatize(token, pos = "n") for token in tokens if token.lower() not in swords]
tokens2 = [token.lower() for token in tokens1 if token.isalnum()]

freq=FreqDist(tokens2)
freq.plot(10)

freq.most_common(3)
```
